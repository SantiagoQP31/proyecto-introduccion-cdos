## Generate Stacked Bar Chart for LATAM

## Gráficos para el Storytelling

```python
# Ensure Year is integer
df['Year'] = df['Year'].astype(int)

# Factors list
factors = [
    'Explained by: Log GDP per capita',
    'Explained by: Social support',
    'Explained by: Healthy life expectancy',
    'Explained by: Freedom to make life choices',
    'Explained by: Generosity',
    'Explained by: Perceptions of corruption'
]
factor_labels = ['PIB', 'Soporte Social', 'Expectativa de Vida', 'Libertad', 'Generosidad', 'Corrupción']

# ==========================================
# Gráfico 1: Scatter Plot - Riqueza vs Felicidad (2024)
# ==========================================
df_2024 = df[df['Year'] == 2024].copy()
latam_countries = [
    'Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Costa Rica',
    'Dominican Republic', 'Ecuador', 'El Salvador', 'Guatemala', 'Honduras',
    'Mexico', 'Nicaragua', 'Panama', 'Paraguay', 'Peru', 'Uruguay', 'Venezuela'
]

plt.figure(figsize=(10, 6))
# All other countries
df_others = df_2024[~df_2024['Country name'].isin(latam_countries)]
plt.scatter(df_others['Explained by: Log GDP per capita'], df_others['Ladder score'],
            color='lightgrey', alpha=0.6, label='Resto del Mundo')

# Latam
df_latam = df_2024[df_2024['Country name'].isin(latam_countries) & (df_2024['Country name'] != 'Colombia')]
plt.scatter(df_latam['Explained by: Log GDP per capita'], df_latam['Ladder score'],
            color='skyblue', alpha=0.9, label='Latinoamérica')

# Colombia
df_col_2024 = df_2024[df_2024['Country name'] == 'Colombia']
if not df_col_2024.empty:
    plt.scatter(df_col_2024['Explained by: Log GDP per capita'], df_col_2024['Ladder score'],
                color='gold', edgecolor='black', s=150, label='Colombia', zorder=5)

plt.title('Riqueza vs Felicidad: ¿Dónde está Colombia en el Mundo? (2024)', fontsize=14, fontweight='bold')
plt.xlabel('Puntaje aportado por PIB per Cápita', fontsize=12)
plt.ylabel('Puntaje de Felicidad (Ladder Score)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('scatter_wealth_vs_happiness.png', dpi=300)
plt.close()

# ==========================================
# Gráfico 2: Stacked Area Chart - Composición histórica (Colombia)
# ==========================================
df_col_all = df[df['Country name'] == 'Colombia'].sort_values('Year').copy()
# Fill NaN if any
df_col_all = df_col_all.fillna(0)

plt.figure(figsize=(12, 7))
plt.stackplot(df_col_all['Year'],
              df_col_all[factors[0]], df_col_all[factors[1]], df_col_all[factors[2]],
              df_col_all[factors[3]], df_col_all[factors[4]], df_col_all[factors[5]],
              df_col_all['Dystopia + residual'],
              labels=factor_labels + ['Distopía + Residual'],
              colors=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2'],
              alpha=0.85)

plt.title('De qué está hecha la felicidad en Colombia (Histórico)', fontsize=14, fontweight='bold')
plt.xlabel('Año', fontsize=12)
plt.ylabel('Composición del Puntaje de Felicidad', fontsize=12)
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0.)
plt.xticks(df_col_all['Year'].astype(int))
plt.margins(x=0)
plt.tight_layout()
plt.savefig('colombia_happiness_composition_area.png', dpi=300)
plt.close()

# ==========================================
# Gráfico 3: Bar Chart Comparativo - Colombia vs Finlandia (2024)
# ==========================================
df_finland_2024 = df_2024[df_2024['Country name'] == 'Finland']

if not df_col_2024.empty and not df_finland_2024.empty:
    col_values = df_col_2024[factors].values[0]
    fin_values = df_finland_2024[factors].values[0]

    x = np.arange(len(factor_labels))
    width = 0.35

    plt.figure(figsize=(10, 6))
    plt.barh(x - width/2, col_values, width, label='Colombia', color='gold', edgecolor='black')
    plt.barh(x + width/2, fin_values, width, label='Finlandia (Top 1)', color='cornflowerblue', edgecolor='black')

    plt.title('Colombia vs El País Más Feliz (Finlandia - 2024)', fontsize=14, fontweight='bold')
    plt.xlabel('Puntaje Aportado a la Felicidad', fontsize=12)
    plt.yticks(x, factor_labels)
    plt.legend()
    plt.grid(axis='x', linestyle='--', alpha=0.5)
    plt.gca().invert_yaxis() # Highest factor at top
    plt.tight_layout()
    plt.savefig('colombia_vs_finland_bar.png', dpi=300)
    plt.close()

print("Graphs generated successfully.")
```
